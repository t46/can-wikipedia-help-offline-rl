```python
epoch = 40
model_name = 'igpt'
env_name = 'hopper'
dataset_name = 'medium'

mi_dict = np.load(f'./results/mi_np_{epoch}_{model_name}_{env_name}_{dataset_name}_666.npy', allow_pickle=True).item()

state_mi_mean = []
state_mi_std = []
action_mi_mean = []
action_mi_std = []
for block_id, mi_lists in enumerate(mi_dict.values()):
    state_mi_mean.append(np.nanmean(mi_lists[0]))
    state_mi_std.append(np.nanstd(mi_lists[0]))
    action_mi_mean.append(np.nanmean(mi_lists[1]))
    action_mi_std.append(np.nanstd(mi_lists[1]))

state_mi_mean = np.array(state_mi_mean)
state_mi_std = np.array(state_mi_std)
action_mi_mean = np.array(action_mi_mean)
action_mi_std = np.array(action_mi_std)

plt.fill_between(np.arange(len(state_mi_mean)), state_mi_mean - state_mi_std, state_mi_mean + state_mi_std, alpha=0.2, color=(0.372, 0.537, 0.537))
plt.plot(state_mi_mean, color=(0.372, 0.537, 0.537), label=r'$\hat{I}(X, T)$')
plt.scatter(np.arange(len(state_mi_mean)), state_mi_mean, color=(0.372, 0.537, 0.537))

plt.fill_between(np.arange(len(action_mi_mean)), action_mi_mean - action_mi_std, action_mi_mean + action_mi_std, alpha=0.2, color=(0.627, 0.352, 0.470))
plt.plot(action_mi_mean, color=(0.627, 0.352, 0.470), label=r'$\hat{I}(Y, T)$')
plt.scatter(np.arange(len(action_mi_mean)), action_mi_mean, color=(0.627, 0.352, 0.470))

plt.title('Estimated Mutual Information')
plt.xlabel('Transformer Block ID')
plt.ylabel(r'Mean of $\hat{I}$')
plt.ylim(-0.5, 4.5)
plt.legend()
plt.tight_layout()
plt.savefig(f'figs/mi_{epoch}_{model_name}_{env_name}_{dataset_name}_666.pdf')
plt.show()
```
