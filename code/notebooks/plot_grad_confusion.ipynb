```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("ticks")
sns.set_context("paper", 1.5, {"lines.linewidth": 2})
```

```python
epoch = 1
env_name = 'halfcheetah'
dataset_name = 'medium'
seed = 666
reward_state_action = 'state'
model_name_list = ['gpt2', 'igpt', 'dt']

gradcossims_list = []
for model_name in model_name_list:
    gradcossim = np.load(f'results/gradcossim_{epoch}_{model_name}_{env_name}_{dataset_name}_{seed}_{reward_state_action}.npy')
    gradcossims_list.append(gradcossim)

model_name_label = ['GPT2', 'iGPT', 'Random Init']
colors = [(0.372, 0.537, 0.537), (0.627, 0.352, 0.470), (0.733, 0.737, 0.870)]
my_palette = sns.color_palette(colors)
sns.boxplot(data=gradcossims_list, palette=my_palette)  # "PuBuGn_r"
plt.xticks(np.arange(3), model_name_label)
plt.ylabel(r'$cossim(\nabla_{\theta}\ell_1, \nabla_{\theta}\ell_2)$')
# plt.title('Gradient Confusion')
# plt.savefig(f'figs/gradcossim_{epoch}_gpt2_igpt_dt_{env_name}_{dataset_name}_{seed}_{reward_state_action}.pdf')
plt.show()
```

```python
epoch = 1
env_name = 'walker2d'
dataset_name = 'medium'
seed = 666
reward_state_action = 'state'
model_name_list = ['gpt2', 'igpt', 'dt']

min_gradcossims_list = []
for model_name in model_name_list:
    min_gradcossim = np.min(np.load(f'results/gradcossim_{epoch}_{model_name}_{env_name}_{dataset_name}_{seed}_{reward_state_action}.npy'))
    min_gradcossims_list.append([min_gradcossim])

model_name_label = ['GPT2', 'iGPT', 'Random Init']
colors = [(0.372, 0.537, 0.537), (0.627, 0.352, 0.470), (0.733, 0.737, 0.870)]
my_palette = sns.color_palette(colors)
sns.barplot(data=min_gradcossims_list, palette=my_palette)  # "PuBuGn_r"
plt.xticks(np.arange(3), model_name_label)
plt.ylabel(r'Min of $cossim(\nabla_{\theta}\ell_1, \nabla_{\theta}\ell_2)$')
# plt.title('Gradient Confusion')
plt.savefig(f'figs/mingradcossim_{epoch}_gpt2_igpt_dt_{env_name}_{dataset_name}_{seed}_{reward_state_action}.pdf')
plt.show()
```
